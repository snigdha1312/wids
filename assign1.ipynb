```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in ['age','study_hours','class_attendance','sleep_hours','exam_score']:
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
def cap_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return col.clip(lower, upper)

for col in ['study_hours','sleep_hours','exam_score','class_attendance']:
    df[col] = cap_outliers(df[col])
```

```python
df['gender'] = df['gender'].map({'male':0, 'female':1})
df['internet_access'] = df['internet_access'].map({'no':0, 'yes':1})
```

```python
df['sleep_quality'] = df['sleep_quality'].map({'poor':1,'average':2,'good':3})
df['facility_rating'] = df['facility_rating'].map({'low':1,'medium':2,'high':3})
df['exam_difficulty'] = df['exam_difficulty'].map({'easy':1,'moderate':2,'hard':3})
```
